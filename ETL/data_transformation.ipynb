# Extract CSVs into DataFrames

## Cost of Living - 2022

```python
# import the csv to df
CostLiving2022_data = pd.read_csv("../Resources/csv/Cost_of_Living_Index_2022.csv")
CostLiving2022_data['Year'] = 2022
CostLiving2022_data.head()
```

```python
#Drop the column  Groceries Index and Restaurant Price Index 
CostLiving2022_data = CostLiving2022_data.drop("Groceries Index", axis=1)
CostLiving2022_data = CostLiving2022_data.drop("Restaurant Price Index", axis=1)
CostLiving2022_data = CostLiving2022_data.drop("Rank", axis=1)
CostLiving2022_data.head()
```

```python
# Most 10 expensive country - 2022
CL2022_Expensive = CostLiving2022_data.nlargest(10, 'Cost of Living Index')
CL2022_Expensive
```

```python
#This database is based on New York City.
#We can see Bermuda is 46% more expensive to live than New York, Groceries and Restaurant increase the price. 
#Also, we can see that based on Local Purchasing Power Index is lower than new york
#Switerland is 23% more expensive than NY. Rent index is half than new york. However the price of food it seems to be 
#around 25% more expensive
```

```python
#Cheapest Country to Live - 2022
CL2022_Cheap = CostLiving2022_data.nsmallest(10, 'Cost of Living Index')
CL2022_Cheap
```

```python
#MEAN Cost 2022
MeanCost2022 = CostLiving2022_data['Cost of Living Index'].mean()
MeanCost2022
```

```python
Highest_Purchasing_Power2022 = CostLiving2022_data.nlargest(10, 'Local Purchasing Power Index')
Highest_Purchasing_Power2022
```

```python
Lowest_Purchasing_Power2022 = CostLiving2022_data.nsmallest(10, 'Local Purchasing Power Index')
Lowest_Purchasing_Power2022
```

```python
#Most expensive Rent
Rent_Most_Expensive = CostLiving2022_data.nlargest(10,'Rent Index')
Rent_Most_Expensive
```

## Cost of Living - 2020

```python
# import the csv to df
CostLiving2020_data = pd.read_csv("../Resources/csv/cost of living 2020.csv")
CostLiving2020_data['Year'] = 2020
CostLiving2020_data.head()
```

```python
#Drop the column  Groceries Index and Restaurant Price Index 
CostLiving2020_data = CostLiving2020_data.drop("Groceries Index", axis=1)
CostLiving2020_data = CostLiving2020_data.drop("Restaurant Price Index", axis=1)
CostLiving2020_data = CostLiving2020_data.drop("Rank 2020", axis=1)
CostLiving2020_data.head()
```

```python
#Drop the column  McMeal($) and Unnamed: 9  
CostLiving2020_data = CostLiving2020_data.drop("McMeal($)", axis=1)
CostLiving2020_data = CostLiving2020_data.drop("Unnamed: 9", axis=1)

# Rename Rank Column
CostLiving2020_data = CostLiving2020_data.rename(columns = {'Rank 2020': 'Rank'})

CostLiving2020_data.head()
```

```python
# Most 10 expensive Country
CL2020_Expensive = CostLiving2020_data.nlargest(10, 'Cost of Living Index')
CL2020_Expensive
```

```python
#Cheapest Country to Live - 2020
CL2020_Cheap = CostLiving2020_data.nsmallest(10, 'Cost of Living Index')
CL2020_Cheap
```

```python
#MEAN Cost 2020
MeanCost2020 = CostLiving2020_data['Cost of Living Index'].mean()
MeanCost2020
```

```python
Purchasing_Power2020 = CostLiving2020_data.nlargest(10, 'Local Purchasing Power Index')
Purchasing_Power2020
```

```python
#Most expensive Rent
Rent_Most_Expensive = CostLiving2020_data.nlargest(10,'Rent Index')
Rent_Most_Expensive
```
